### TreeMape



```python
from pyecharts import TreeMap

data = [
    {
        "value": 40,
        "name": "A",
    },
    {
        "value": 180,
        "name": "B",
        "children": [
            {
                "value": 76,
                "name": "B.children",
                "children": [
                    {
                        "value": 12,
                        "name": "fasfsa",
                    },
                    {
                        "value": 28,
                        "name": "B.children.b",
                    },
                    {
                        "value": 20,
                        "name": "B.children.c",
                    },
                    {
                        "value": 16,
                        "name": "B.children.d",
                    }]
            }]},
    {
        "value": 200,
        "name": "C",

        "children": [
            {
                "value": 12,
                "name": "C.children.a",
            },
            {
                "value": 28,
                "name": "C.children.b",
            },
            {
                "value": 20,
                "name": "C.children.c",
            },
            {
                "value": 200,
                "name": "C.children.c1",
            },
            {
                "value": 16,
                "name": "C.children.d",
            }]
    }
]

treemap = TreeMap("-", width=1200, height=600)
treemap.add("", data, is_label_show=True, label_pos='inside')
treemap.render()
```
